```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "http://quotes.toscrape.com"
```

```python
url = "http://quotes.toscrape.com"
```

```python
response = requests.get(url)
```

```python
print(response.status_code)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
print(soup.title.text)
```

```python
quotes = soup.find_all("div", class_="quote")
```

```python
print(len(quotes))
```

```python
quote_texts = []

for quote in quotes:
    text = quote.find("span", class_="text").text
    quote_texts.append(text)
```

```python
print(quote_texts)
```

```python
authors = []

for quote in quotes:
    author = quote.find("small", class_="author").text
    authors.append(author)
```

```python
print(authors)
```

```python
tags = []

for quote in quotes:
    tag = ", ".join([t.text for t in quote.find_all("a", class_="tag")])
    tags.append(tag)
```

```python
print(tags)
```

```python
df = pd.DataFrame({
    "Quote": quote_texts,
    "Author": authors,
    "Tags": tags
})
```

```python
df.head()
```

```python
df.to_csv("quotes_dataset.csv", index=False)

print("Dataset saved successfully!")
```
